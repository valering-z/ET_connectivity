```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
import matplotlib as mpl

mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42
mpl.rcParams['font.family'] = 'Arial'
```

```python
col_stats = pd.read_csv('Column-L5ET-recip_stats.csv')
ext_stats = pd.read_csv('Extended-L5ET-recip-synapse_stats.csv')
```

```python

# Create a figure
fig = plt.figure(figsize=(5, 14)) 


gs = gridspec.GridSpec(5, 1, height_ratios=[0.8,0.2,1.6,0.2,1.6], hspace=0.4)


ax1 = fig.add_subplot(gs[0, 0])  # First row
for pos in ['right', 'top']:
    ax1.spines[pos].set_visible(False)

sns.violinplot(ax=ax1, data=col_stats[['Group 1','Group 2']], inner=None, density_norm='width', orient='y', linewidth=0, cut=0)
sns.stripplot(ax=ax1, data=col_stats[['Group 1','Group 2']], orient='y', color='black', size=4, jitter=0.3)

ax1.set_title('a', x=-0.35, y=1.1, loc='left', fontsize=16, fontname='Arial')
ax1.set_xlabel('Reciprocal rate (%)', fontsize=13, fontname='Arial')

ax2 = fig.add_subplot(gs[2:3, 0]) 
for pos in ['right', 'top']:
    ax2.spines[pos].set_visible(False)

sns.scatterplot(ax=ax2, data=ext_stats, x='sum_size_lv1', y='sum_size_r', color='slategrey',  edgecolor='white', s=30, alpha=1) 

ax2.set_title('b', x=-0.35, y=1.05, loc='left', fontsize=16, fontname='Arial')
ax2.set_xlabel('Synapse size of reciprocal \nET to inhibitory to ET connections \n(voxels)', fontsize=13, fontname='Arial')
ax2.set_ylabel('Synapse size of reciprocal \ninhibitory to ET connections \n(voxels)', fontsize=13, fontname='Arial')


ax3 = fig.add_subplot(gs[4:, 0]) 
for pos in ['right', 'top']:
    ax3.spines[pos].set_visible(False)

sns.scatterplot(ax=ax3, data=ext_stats, x='n_syn_lv1', y='n_syn_r', color='slategrey', edgecolor='white', s=30, alpha=1) 

ax3.set_title('c', x=-0.35, y=1.05, loc='left', fontsize=16, fontname='Arial')
ax3.set_xlabel('Synapse number of reciprocal \nET to inhibitory to ET connections', fontsize=13, fontname='Arial')
ax3.set_ylabel('Synapse number of reciprocal \ninhibitory to ET connections', fontsize=13, fontname='Arial')


plt.savefig('ET-extended-figure-12.svg', dpi=300,  bbox_inches='tight')
```
